```python
# Importing libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
# Importing diabetes dataset
diabetes_df = pd.read_csv('diabetes_datasets/pima_diabetes.csv')
diabetes_df
```

```python
# Seperate the features, X,  from the target variable, y
y = diabetes_df['Outcome']
X = diabetes_df.drop(columns='Outcome')
```

```python
# Preview the features
X.head()
```

```python
#preview the few entries for the target variable
y[:5]
```

```python
# Encode the categorical variables using get_dummies
X = pd.get_dummies(X)
X.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Scale the data using StandardScalar
# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Statistical summary
X.describe().T
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs',
                                max_iter=200,
                                random_state=1)
classifier
```

```python
#Fit model using training data
classifier.fit(X_train, y_train)
```

```python
#Score the modle using the test data
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
#Make predictions
predictions = classifier.predict(X_test)
results = pd.DataFrame({"Prediction": predictions, "Actual": y_test}).reset_index(drop=True)
results.head(10)
```

```python
#Calculate the Accuracy score
from sklearn.metrics import accuracy_score
# Display the accuracy score for the test dataset.
accuracy_score(y_test, predictions)
```

```python
#Export the predictions to csv
results.to_csv('linreg_predictions.csv', index=False)
```
